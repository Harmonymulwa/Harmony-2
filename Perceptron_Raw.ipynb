```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

## Training Data set

```python
Assign_Train = [10,10,80,30]
Exam_Train   = [80,10,10,30]
Result      =  [1,  0, 1, 0]
```

## Ploting Train Data set.
ploting data points using seaborn (Seaborn requires dataframe)

```python
df = pd.DataFrame(
    {'x1': Assign_Train,
     'x2': Exam_Train,
     'y': Result
    })

plt.figure(0)    
sns.lmplot("x1", "x2", data=df, hue='y', fit_reg=False, markers=["o", "s"])
```

## Activation Function

```python
def step_function(x):
    if x<0:
        return 0
    else:
        return 1

def sigmoid_function(x):
  return 1 / (1 + np.exp(-x))
```

## Parameter Initialization

```python
#w = np.random.rand(2)
w = np.array([1,-1])
errors = [] 
eta = .90   # learning rate
epoch = 1000
b = 0
```

## Learning

```python
for i in range(epoch):
    for x,y in zip(zip(Assign_Train,Exam_Train),Result):
      # u = np.dot(x , w) +b
        u = sum(x*w) + b
        
     #   error = y - sigmoid_function(u) 
        error = y - step_function(u) 
      
        errors.append(error) 
        for index, value in enumerate(x):
            w[index] += eta * error * value
            b += eta*error
   
        # produce all decision boundaries
            a = [0,-b/w[1]]
            c = [-b/w[0],0]
            plt.figure(1)
            plt.plot(a,c)
        
            
```

# Ploting Final decision boundary

```python
sns.lmplot("x1", "x2", data=df, hue='y', fit_reg=False, markers=["o", "s"])
a = [0,-b/w[1]]
c = [-b/w[0],0]
plt.plot(a,c)
```

# Ploting errors

```python
plt.figure(3)
plt.ylim([-1,1]) 
plt.plot(errors)
```
